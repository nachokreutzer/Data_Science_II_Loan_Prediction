#Bank Loan Status Dataset

* El análisis del riesgo crediticio es fundamental para las instituciones financieras, ya que les permite identificar clientes potenciales que podrían no cumplir con sus obligaciones de pago. Con la creciente demanda de crédito en el mercado actual, la capacidad de predecir la probabilidad de incumplimiento se ha convertido en una herramienta crucial para los bancos y otros prestamistas. El objetivo de este proyecto es utilizar técnicas de Machine Learning para predecir la aprobación o rechazo de préstamos basándose en factores clave como el historial de crédito, los ingresos anuales y la deuda mensual.

* Este análisis proporcionará a las instituciones financieras una base más sólida para tomar decisiones informadas sobre el otorgamiento de préstamos, lo que podría minimizar las pérdidas por incumplimientos y optimizar el flujo de efectivo. La audiencia beneficiaria de este análisis son los ejecutivos de bancos y los responsables de riesgo, que necesitan mejorar la precisión de sus decisiones basadas en datos para reducir el riesgo y aumentar la rentabilidad.

###Diccionario de datos

Loan ID ---> Identificador único del préstamo

Customer ID ---> Identificador único del cliente

Loan Status ---> Estado del préstamo (Fully Paid: pagado completamente, Charged Off: cancelado por incumplimiento)

Current Loan Amount ---> Monto actual del préstamo solicitado

Term ---> Plazo del préstamo (duración del pago)

Credit Score ---> Puntaje crediticio del solicitante

Annual Income ---> Ingresos anuales del solicitante

Years in current job ---> Años en el empleo actual del solicitante

Home Ownership ---> Estado de propiedad de vivienda (Alquilada, Propia, Hipotecada, etc.)

Purpose ---> Propósito del préstamo (consolidación de deuda, compra de auto, entre otros)

Monthly Debt ---> Deuda mensual del solicitante

Years of Credit History ---> Años de historial crediticio del solicitante

Months since last delinquent ---> Meses desde la última morosidad en pagos

Number of Open Accounts ---> Número de cuentas abiertas del solicitante

Number of Credit Problems ---> Número de problemas crediticios del solicitante

Current Credit Balance ---> Saldo actual del crédito del solicitante

Maximum Open Credit ---> Crédito máximo disponible

Bankruptcies ---> Número de bancarrotas del solicitante

Tax Liens ---> Gravámenes fiscales del solicitante

#Conclusiones Finales


El análisis exploratorio de datos (EDA) realizado sobre el dataset de estado de préstamos bancarios ha revelado varios patrones interesantes que pueden ser útiles para mejorar la toma de decisiones en el proceso de aprobación de préstamos. A continuación, se presentan las conclusiones clave obtenidas del análisis:

* Relación entre puntaje crediticio y estado del préstamo:
Los solicitantes con puntajes crediticios más bajos tienden a tener una mayor probabilidad de que sus préstamos sean cancelados por incumplimiento (Charged Off). Los puntajes crediticios más altos están asociados con una mayor probabilidad de que el préstamo sea completamente pagado (Fully Paid). Esto refuerza la importancia de evaluar cuidadosamente el historial crediticio al tomar decisiones sobre préstamos.

* Impacto de los ingresos anuales:
Aunque los ingresos anuales son un factor relevante en la evaluación crediticia, no se observó una relación clara y directa entre los ingresos y el monto del préstamo. Sin embargo, se identificó una correlación moderada entre los ingresos y la deuda mensual, lo que sugiere que los ingresos más altos permiten asumir mayores responsabilidades financieras.

* Años en el empleo actual:
No se observó una relación fuerte entre los años en el empleo actual y el estado del préstamo. Aunque este factor podría proporcionar estabilidad en los ingresos, no parece ser un determinante clave en la aprobación o cancelación de los préstamos.

* Variables con mayor influencia:
El puntaje crediticio sigue siendo la variable más importante a la hora de determinar el estado del préstamo, seguido por factores como la cantidad de problemas crediticios y el historial crediticio. Estas variables pueden ayudar a los bancos a predecir con mayor precisión qué solicitantes son más propensos a incumplir con sus pagos.

* Características de los solicitantes rechazados:
Los solicitantes cuyo crédito fue cancelado (Charged Off) tienden a tener menores puntajes crediticios, más problemas crediticios registrados y, en algunos casos, ingresos relativamente más bajos. Estas características deben ser consideradas al evaluar el riesgo crediticio.

Recomendaciones
Con base en los resultados obtenidos, se sugiere a las instituciones financieras que:

* Den prioridad al puntaje crediticio y a los problemas crediticios al evaluar nuevos solicitantes.
* Apliquen políticas de mitigación de riesgos, especialmente en aquellos solicitantes con un puntaje crediticio bajo o un historial de problemas financieros.
* Revisen los montos de los préstamos ofrecidos a solicitantes con ingresos muy elevados para evitar sobreendeudamiento, incluso cuando estos solicitantes tienen buenos puntajes crediticios.
